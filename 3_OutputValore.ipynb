## Analisi dei grafici stagionali 2019

Dall'analisi dei grafici relativi alle quattro stagioni del 2019 emerge che:

- Giorni ad alto rischio sanitario sono presenti in tutte le stagioni, non solo in estate o inverno.
- In primavera e autunno, il rischio sanitario Ã¨ spesso legato a improvvisi cali di temperatura.
- In estate, l'ondata di calore Ã¨ costante e aumenta il rischio anche nei giorni non estremi.
- In inverno, il freddo persistente rappresenta il rischio maggiore.

Queste evidenze confermano quanto mostrato dal nostro heatmap: **anche giornate moderatamente calde o fredde possono rappresentare un rischio sanitario significativo**. Questo insight Ã¨ fondamentale per le nostre personae e per la progettazione di un sistema di allerta climatica piÃ¹ efficace.


```python
# Inizializza dizionario per contare giorni a rischio
contatore = {}

# Loop su tutte le stagioni
for season in [1, 2, 3, 4]:

    # Filtra l'anno e la stagione
    df_year = df[(df.period == 2019) & (df.season == season)]

    # Calcola soglia di rischio
    threshold = df_year['mortality_risk_pct'].mean()

    # Trova giorni sopra soglia
    high_risk_days = df_year[df_year['mortality_risk_pct'] > threshold]

    # Conta righe
    count = high_risk_days.shape[0]

    # Salva nel dizionario
    contatore[season] = count

print(contatore)
```

## Analisi dei giorni a rischio sanitario per stagione

Dall'analisi emerge che:

- **Estate (2)** Ã¨ la stagione con il maggior numero di giorni ad alto rischio sanitario (56 giorni), confermando il rischio legato alle ondate di calore.
- Anche l'**inverno (4)** mostra un numero elevato di giorni a rischio (46), coerente con l'effetto negativo del freddo sulla salute pubblica.
- **Primavera (1)** e **autunno (3)** non sono esenti da giorni pericolosi, a dimostrazione che anche le temperature moderate possono rappresentare un rischio sanitario.

Questi risultati sono fondamentali per pianificare misure di prevenzione sanitaria e per rispondere ai bisogni delle nostre personae.


```python
# Calcola popolazione media per stagione
pop_per_season = df[df.period == 2019].groupby('season')[['over65', 'under5']].mean()

# Calcola giornate-persona esposte per ogni stagione
for season, days in contatore.items():
    pop_over65 = pop_per_season.loc[season, 'over65']
    pop_under5 = pop_per_season.loc[season, 'under5']

    exposure_over65 = days * pop_over65
    exposure_under5 = days * pop_under5

    print(f"Stagione {season}: over65 esposti = {exposure_over65:.0f}, under5 esposti = {exposure_under5:.0f}")
```

```python
# Dizionario con i risultati
dataplot = {
    'Stagione': ['Primavera', 'Estate', 'Autunno', 'Inverno'],
    'over65': [8137201, 17525872, 12518480, 14396252],
    'under5': [1267136, 2729216, 1949440, 2241856]
}

df_plot = pd.DataFrame(dataplot)

# Trasforma il dataframe in long format
df_long = df_plot.melt(id_vars='Stagione', value_vars=['over65', 'under5'],
                       var_name='Gruppo', value_name='Giornate_persona')

# Plot
plt.figure(figsize=(10,6))
sns.barplot(
    data=df_long,
    x='Stagione',
    y='Giornate_persona',
    hue='Gruppo'
)

plt.title("Giornate-persona esposte al rischio sanitario per stagione (2019)")
plt.ylabel("Giornate-persona esposte")
plt.xlabel("Stagione")
plt.legend(title='Gruppo di etÃ ')
plt.grid(True)
plt.show()
```

## Interpretazione delle giornate-persona esposte al rischio sanitario

Dal grafico emerge chiaramente che l'**Estate** rappresenta la stagione con il maggior numero di giornate-persona esposte al rischio sanitario, sia per il gruppo `over65` sia per il gruppo `under5`. Tuttavia, anche Inverno, Autunno e Primavera mostrano valori significativi, confermando che il rischio sanitario non Ã¨ limitato solo agli estremi stagionali.

### Collegamento alle nostre personae:

- **Luciana (84 anni)**:  
  L'elevato numero di giornate-persona a rischio in inverno e anche nelle stagioni intermedie sottolinea l'urgenza di un sistema di allerta per proteggere le persone anziane come Luciana, particolarmente vulnerabili alle temperature moderate o fredde.

- **Manuela (31 anni, madre)**:  
  Anche se numericamente inferiore, il gruppo `under5` mostra un'esposizione rilevante, soprattutto in estate. Questo conferma l'importanza per Manuela di sapere quali giorni siano sicuri per portare il figlio all'aperto senza rischi sanitari.

- **Marco (40 anni, operatore sanitario)**:  
  I dati sulle giornate-persona esposte aiutano Marco a pianificare risorse sanitarie e prevenire emergenze, specialmente nelle stagioni piÃ¹ critiche come l'estate e l'inverno.

Queste evidenze dimostrano l'importanza di un sistema predittivo basato sui dati, in grado di fornire allerte mirate per ridurre il rischio sanitario e rispondere ai bisogni delle diverse fasce della popolazione.


```python
# Popolazione media per anno e stagione
pop_per_year_season = df.groupby(['period', 'season'])[['over65', 'under5']].mean()

# Threshold dinamico per ogni anno e stagione
thresholds = df.groupby(['period', 'season'])['mortality_risk_pct'].mean()

# Conta giorni sopra soglia per ogni anno e stagione
contador = {}

for (period, season), threshold_value in thresholds.items():
    mask = (
        (df['period'] == period) &
        (df['season'] == season) &
        (df['mortality_risk_pct'] > threshold_value)
    )
    days = df.loc[mask].shape[0]
    contador[(period, season)] = days

# Calcola giornate-persona esposte
exposure_data = []

for (period, season), days in contador.items():
    pop_over65 = pop_per_year_season.loc[(period, season), 'over65']
    pop_under5 = pop_per_year_season.loc[(period, season), 'under5']

    exposure_over65 = days * pop_over65
    exposure_under5 = days * pop_under5

    exposure_data.append({
        'period': period,
        'season': season,
        'over65': exposure_over65,
        'under5': exposure_under5,
    })

df_exposure = pd.DataFrame(exposure_data)

# Somma giornate-persona per anno e stagione
df_year_season = df_exposure.copy()
df_year_season['total'] = df_year_season['over65'] + df_year_season['under5']

# Plot con colori contrastanti
season_order = [1, 2, 3, 4]
colors = ['#2ca02c', '#d62728', '#ff7f0e', '#1f77b4']

plt.figure(figsize=(12,6))
sns.barplot(
    data=df_year_season,
    x='period',
    y='total',
    hue='season',
    hue_order=season_order,
    palette=colors
)

plt.title("Totale giornate-persona esposte al rischio sanitario per anno, suddiviso per stagione")
plt.ylabel("Giornate-persona esposte")
plt.xlabel("Anno")
plt.legend(
    title='Stagione',
    labels=['Primavera (1)', 'Estate (2)', 'Autunno (3)', 'Inverno (4)']
)
plt.grid(True)
plt.show()
```

## Interpretazione del grafico delle giornate-persona esposte al rischio sanitario per anno e stagione

Dal grafico osserviamo che:

- **Estate (2)** e **Inverno (4)** risultano le stagioni con il numero piÃ¹ alto di giornate-persona esposte al rischio sanitario nella maggior parte degli anni. Questo conferma il peso sia delle ondate di calore estive sia del rischio legato al freddo invernale.

- **Primavera (1)** mostra valori costantemente piÃ¹ bassi rispetto alle altre stagioni, ma comunque significativi, a dimostrazione che anche le temperature moderate possono costituire un rischio sanitario.

- **Autunno (3)** si posiziona in una fascia intermedia, senza raggiungere picchi estremi ma contribuendo in modo stabile al totale delle giornate-persona esposte.

- Non si osserva una crescita costante del rischio sanitario negli anni, ma piuttosto variazioni tra un anno e l'altro, il che evidenzia come il rischio sia fortemente influenzato dalle condizioni climatiche specifiche di ciascun anno.

- Il calo osservato nel 2025 potrebbe essere legato a:
  - una diminuzione effettiva del rischio sanitario,
  - oppure a una parziale mancanza di dati per quell'anno.

Queste informazioni sono fondamentali per pianificare strategie di prevenzione sanitaria mirate e per rispondere alle esigenze delle diverse **personae** coinvolte, in particolare:
- **Luciana** (84 anni), piÃ¹ esposta ai rischi invernali e alle temperature estreme.
- **Manuela** e il figlio piccolo, maggiormente vulnerabili nelle ondate di calore estive.

Il grafico dimostra la necessitÃ  di monitorare **tutte le stagioni** e non solo i periodi di temperature estreme.


```python
# Trasforma in long format
df_long = df_year_season.melt(
    id_vars=['period', 'season'],
    value_vars=['over65', 'under5'],
    var_name='Gruppo',
    value_name='Giornate_persona'
)

season_order = [1, 2, 3, 4]
season_labels = ['Primavera (1)', 'Estate (2)', 'Autunno (3)', 'Inverno (4)']
palette_groups = {'over65': '#1f77b4', 'under5': '#ff7f0e'}

g = sns.catplot(
    data=df_long,
    kind='bar',
    x='period',
    y='Giornate_persona',
    hue='Gruppo',
    col='season',
    col_order=season_order,
    palette=palette_groups,
    height=5,
    aspect=1.2,
    legend_out=True
)

g.set_axis_labels("Anno", "Giornate-persona esposte")
g.set_titles("{col_name}")
for ax, label in zip(g.axes.flat, season_labels):
    ax.set_title(label)

plt.subplots_adjust(top=0.85)
g.fig.suptitle("Giornate-persona esposte al rischio sanitario per anno, suddivise per gruppo e stagione")

plt.show()
```
